```python
import pandas as pd
import numpy as np
```

```python
# For Excel
df = pd.read_excel('/content/API_SP.POP.TOTL_DS2_en_excel_v2_73799.xls', sheet_name='Data', skiprows=3)
```

```python
df.head()         # View first 5 rows
df.info()         # Column types & non-null counts
df.describe()     # Summary stats (only for numerical columns)
```

```python
df.isnull().sum()
```

```python
# Drop rows with any missing value
df_clean = df.dropna()
```

```python
# Print the columns to verify their names
print(df.columns)

# Drop columns. If 'Unnamed: 66' is not in the output of the print statement above, remove it from the list.
df = df.drop(columns=['Unnamed: 66', 'Indicator Name', 'Indicator Code'], errors='ignore')
```

```python
df.duplicated().sum()     # Count duplicates
df = df.drop_duplicates()
```

```python
df.rename(columns={'Country Name': 'Country'}, inplace=True)
```

```python
df['2020'] = pd.to_numeric(df['2020'], errors='coerce')  # convert to float/int
```

```python
df.to_csv('cleaned_data.csv', index=False)
#save data
```

```python
year = '2020'
df_year = df[['Country', year]].dropna()
```
